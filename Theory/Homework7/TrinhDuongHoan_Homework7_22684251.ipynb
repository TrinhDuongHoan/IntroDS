Chú thích :
+ Number of times pregnant : Số lần mang thai
+ Plasma glucose concentration : Nồng độ glucose huyết tương 2 giờ trong xét nghiệm dung nạp glucose đường uống
+ Diastolic blood pressure (mm Hg) : Huyết áp tâm trương 
+ Triceps skin fold thickness (mm) : Độ dày nếp gấp da cơ tam đầu (mm)
+ 2+Hour serum insulin (mu U/ml) : Insulin huyết thanh 2 giờ (mu U/ml)
+ Body mass index : Chỉ số khối cơ thể
+ Diabetes pedigree function : Chức năng phả hệ bệnh tiểu đường
+ Age : Tuổi
+ Class variable : Lớp biến biểu diễn có mang bệnh tiểu đường hay không

```python
# Preprocess
file_name = 'pima-indians-diabetes-Full_Description.csv' 
col_1 = 'Number of times pregnant' 
col_2 = 'Plasma glucose concentration'
col_3 = 'Diastolic blood pressure (mm Hg)'
col_4 = 'Triceps skin fold thickness (mm)'
col_5 = '2-Hour serum insulin (mu U/ml)'
col_6 = 'Body mass index'
col_7 = 'Diabetes pedigree function'
col_8 = 'Age'
col_9 = 'Class'

name_columns = [col_1,col_2,col_3,col_4,col_5,col_6,col_7,col_8,col_9]
```

```python
data  = pd.read_csv(file_name, header= None, names=name_columns)
data
```

```python
df = pd.DataFrame(data)
df.rename(columns={'Class' : 'Có bệnh/Không có'}, inplace= True)
df.rename(columns={'Number of times pregnant' : 'Số lượt mang thai'}, inplace=True)
new_data = pd.crosstab(df['Số lượt mang thai'],df['Có bệnh/Không có'], margins=True, margins_name='Tổng')
new_data = pd.DataFrame(new_data)

new_data.rename(columns={0 : 'Có bệnh', 1 : 'Không có'}, inplace= True)
new_data
```

- 1.6

```python
# Lưu dữ liệu vào file
new_data.to_csv('new_data.csv')
```

```python
# Biểu đồ plot
dt = pd.read_csv('new_data.csv')
dt = pd.DataFrame(dt)
dt = dt.iloc[:-1]

plt.plot(dt['Số lượt mang thai'], dt['Có bệnh'], label = 'Có bệnh')
plt.plot(dt['Số lượt mang thai'], dt['Không có'], label = 'Không bệnh')

plt.title('Thống kê về tình trạng người mang thai bị bệnh tiểu đường')
plt.xlabel('Số lượt mang thai')
plt.ylabel('Có bệnh/Không bệnh')
plt.legend()

plt.savefig('plot.png')

plt.show()
```

```python
# Biểu đồ cột

plt.bar(dt['Số lượt mang thai'], dt['Có bệnh'], label = 'Có bệnh')
plt.bar(dt['Số lượt mang thai'], dt['Không có'], label = 'Không bệnh')

plt.title('Thống kê về tình trạng người mang thai bị bệnh tiểu đường')
plt.xlabel('Số lượt mang thai')
plt.ylabel('Có bệnh/Không bệnh')

plt.legend()
plt.savefig('bar.png')
plt.show()
```

- 1.7

```python
age_dis = data.groupby(['Age', 'Class']).size().unstack()
age_dis = age_dis.fillna(0)
age_dis.head()
```

Nhận xét : Bệnh tiểu đường thường bị mắc ở những người trẻ , có độ tuổi vào khoảng 21-40

- 1.8

```python
# Vẽ biểu đồ
plt.figure(figsize=(12,6))
plt.plot(age_dis)
plt.title('Phân bố của bệnh tiểu đường/không có bệnh tiểu đường theo độ tuổi')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(['Không bệnh','Có bệnh'])
plt.show()
```

```python
# mối tương quan giữa số lần mang thai và bệnh tiểu đường theo độ đo Pearson 
data['Number of times pregnant'].corr(data['Class'])
```

- 1.10

```python
# mỗi liên quan giữa các đặc trưng khác với bệnh tiểu đường
correlations = data.iloc[:,:-1].apply(lambda col : col.corr(data['Class']))
correlations
```

Đặc trưng có ảnh hưởng nhất tới bệnh tiểu đường là Plasma glucose concentration. Vì độ tương quan Pearson đối với Class là cao nhất và điều này rất đúng với tình hình thực tế, những người bị bệnh tiểu đường thường có nồng độ đường trong máu khá cao
