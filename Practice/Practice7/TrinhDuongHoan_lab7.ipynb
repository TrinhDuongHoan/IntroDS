```python
tkb = []

# lấy thời gian
thead = driver.find_element(By.TAG_NAME,'thead')
time = [i.text.replace('\n',' ') for i in thead.find_elements(By.TAG_NAME,'th')]

# lấy thời khóa biểu
tbody = driver.find_element(By.TAG_NAME,'tbody')
tr = tbody.find_elements(By.TAG_NAME,'tr')


for i in tr:
    td = i.find_elements(By.TAG_NAME,'td')
    buoi = td[0].text
    cnt = 0
    for j in td:
        lichs = j.find_elements(By.TAG_NAME,'div')
        for k in lichs:
            mh = k.find_element(By.TAG_NAME,'a').text
            elms = k.find_elements(By.TAG_NAME,'p')
            tkb.append(f'Ngày tháng : {time[cnt]} \nThời gian : {elms[1].text}\nMôn học : {mh}\nPhòng học : {elms[2].text}')
        cnt += 1
print(('\n'+'-'*40+'\n').join(tkb))
```

```python
import urllib.request
from bs4 import BeautifulSoup

# URL của trang web cần cào dữ liệu
url = 'file:///D:/lichj%20theo%20tuan/L%E1%BB%8Bch%20theo%20tu%E1%BA%A7n.html'
response = urllib.request.urlopen(url)
html_content = response.read()

soup = BeautifulSoup(html_content, 'html.parser')
for br_tag in soup.find_all('br'):
    br_tag.replace_with(', ')
thead = soup.find('thead')
tbody = soup.find('tbody')

time = [i.text for i in thead.find_all('th')]

tkb = []

for tr in tbody.find_all('tr'):
    td = tr.find_all('td')
    buoi = td[0].get_text(strip=True)
    cnt = 0
    for j in td:
        lichs = j.find_all('div')
        for k in lichs:
            mh = k.find('a').text
            elms = k.find_all('p')
#             for i in elms:
#                 print(i.get_text(strip = True))
            
            tkb.append(f'Ngày tháng : {time[cnt]} \nThời gian : {elms[1].get_text(strip = True)}\nMôn học : {mh}')
        cnt += 1
print(('\n' + '-' * 40 + '\n').join(tkb))
```
