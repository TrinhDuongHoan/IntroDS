```python
from abc import ABC , abstractmethod
import random
from datetime import datetime
```

```python
class Animal(ABC):
  def __init__(self, picture : None, food : None , hunger : None, boundaries : None, location : None) -> None:
    try :
      data = [picture, food, hunger, boundaries,location]
      for i in range(len(data)) :
        if data[i] is None :
          data[i] = 'Ch∆∞a x√°c ƒë·ªãnh'
      if not isinstance(data[0], str) :
        raise Exception ('Picture ph·∫£i l√† ki·ªÉu chu·ªói')
      else :
        self.picture = data[0]
      if not isinstance(data[1], str) :
        raise Exception ('Food ph·∫£i l√† ki·ªÉu chu·ªói')
      else :
        self.food = data[1]
      if not isinstance(data[2], bool) :
        raise Exception ('Hunger ph·∫£i mang gi√° tr·ªã True/False')
      else :
        self.hunger = hunger
      if not isinstance(data[3], str) :
        raise Exception ('Boundaries ph·∫£i l√† ki·ªÉu chu·ªói')
      else :
        self.boundaries = boundaries
      if not isinstance(data[4], str) :
        raise Exception ('Location ph·∫£i l√† ki·ªÉu chu·ªói')
      else :
        self.location = location
    except Exception as E :
      print(E)
  @abstractmethod
  def makeNoise() :
    pass
  @abstractmethod
  def eat() :
    pass
  @abstractmethod
  def roam(self):
    pass
  def sleep(self) :
    return random.choice(['ƒêang ng·ªß','ƒêang th·ª©c'])
  def __str__(self) :
    return f'Picture: {self.picture}, Food: {self.food}, Hunger: {self.hunger}, Boundaries: {self.boundaries}, Location: {self.location}'
```

```python
class Feline(Animal) :
    def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
        super().__init__(picture, food, hunger, boundaries, location)
    
    def roam(self) :
        return random.choice(['ƒêang ƒëi d·∫°o','ƒêang n·∫±m ngh·ªâ'])
```

```python
class Canine(Animal) :
    def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
        super().__init__(picture, food, hunger, boundaries, location)

    def roam(self) :
        return random.choice(['ƒêang ƒëi d·∫°o','ƒêang n·∫±m ngh·ªâ'])
```

```python
class Lion(Feline) :
  def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
    super().__init__(picture, food, hunger, boundaries, location)
  def makeNoise(self):
    print("S∆∞ t·ª≠ ƒëang k√™u")
  def eat(self) :
    if self.food == 'Ch∆∞a x√°c ƒë·ªãnh' :
      print('S∆∞ t·ª≠ ƒëang kh√¥ng ƒÉn g√¨')
    else :
      print(f'S∆∞ t·ª≠ ƒëang ƒÉn {self.food}')
  def __str__(self):
    return f'Name: S∆∞ t·ª≠, '+ super().__str__()
```

```python
class Person :
    def __init__(self, name : None , birthday : None) -> None:
        try :
            if isinstance(name, str) :
                self.name = name 
            else :
                raise Exception('Name ph·∫£i l√† ki·ªÉu chu·ªói')
            datetime.strptime(birthday, '%d-%m-%Y')
            self.birthday = birthday
        except ValueError :
            print('Ng√†y kh√¥ng h·ª£p l·ªá')
        except Exception as E :
            print(E)
    
    def getName(self) :
        return self.name
    
    def __str__(self) -> str:
        return f'Name : {self.name} , Birtday : {self.birthday}'
```

```python
class Employee (Person) :
    def __init__(self, name: None, birthday: None, salary : None) -> None:
        try :
            super().__init__(name, birthday)
            if isinstance(salary,float) or isinstance(salary,int) or salary > 0 :
                self.salary = salary
            else :
                raise Exception('L∆∞∆°ng kh√¥ng h·ª£p l·ªá')
        except Exception as E :
            print(E)

    def getSalary(self) :
        return self.salary

    def __str__(self) -> str:
        return super().__str__() + f', Salary : {self.salary}'
```

```python
class Manager(Employee) :
    def __init__(self, name: None, birthday: None, salary: None, assistant : None) -> None:
        try :
            super().__init__(name, birthday, salary)
            if isinstance(assistant,Employee) :
                self.assistant = assistant
            else :
                raise Exception('Ki·ªÉu c·ªßa assistant kh√¥ng ph√π h·ª£p')
        except Exception as E:
            print(E)
    
    def setAssistant(self, assistant : None) :
        try :
            if isinstance(assistant,Employee) :
                self.assistant = assistant
            else :
                raise Exception('Ki·ªÉu c·ªßa assistant kh√¥ng ph√π h·ª£p')
        except Exception as E :
            print(E)
    
    def __str__(self) -> str:
        return super().__str__() + f', Assistant : {self.assistant.name}'
```

```python
# ƒê·ªãnh nghƒ©a interface Expression
class Expression:
    def toString(self):
        pass

    def evaluate(self):
        pass

# ƒê·ªãnh nghƒ©a interface BinaryExpression k·∫ø th·ª´a t·ª´ Expression
class BinaryExpression(Expression):
    def left(self):
        pass

    def right(self):
        pass

# C√†i ƒë·∫∑t l·ªõp Numeral
class Numeral(Expression):
    def __init__(self, value):
        self.value = value

    def toString(self):
        return str(self.value)

    def evaluate(self):
        return self.value

# C√†i ƒë·∫∑t l·ªõp Square
class Square(Expression):
    def __init__(self, expression):
        self.expression = expression

    def toString(self):
        return f"({self.expression.toString()})^2"

    def evaluate(self):
        return self.expression.evaluate() ** 2

# C√†i ƒë·∫∑t l·ªõp Addition
class Addition(BinaryExpression):
    def __init__(self, left, right):
        self.left = left
        self.right = right

    def toString(self):
        return f"({self.left.toString()} + {self.right.toString()})"

    def evaluate(self):
        return self.left.evaluate() + self.right.evaluate()

# V√≠ d·ª• s·ª≠ d·ª•ng c√°c l·ªõp
num1 = Numeral(3)
num2 = Numeral(4)
sum_expression = Addition(num1, num2)
square_expression = Square(sum_expression)

print(sum_expression.toString())  # Output: (3 + 4)
print(sum_expression.evaluate())  # Output: 7
print(square_expression.toString())  # Output: ((3 + 4))^2
print(square_expression.evaluate())  # Output: 49
```
