```python
from abc import ABC , abstractmethod
import random
from datetime import datetime
```

```python
class Animal(ABC):
  def __init__(self, picture : None, food : None , hunger : None, boundaries : None, location : None) -> None:
    try :
      data = [picture, food, hunger, boundaries,location]
      for i in range(len(data)) :
        if data[i] is None :
          data[i] = 'Chưa xác định'
      if not isinstance(data[0], str) :
        raise Exception ('Picture phải là kiểu chuỗi')
      else :
        self.picture = data[0]
      if not isinstance(data[1], str) :
        raise Exception ('Food phải là kiểu chuỗi')
      else :
        self.food = data[1]
      if not isinstance(data[2], bool) :
        raise Exception ('Hunger phải mang giá trị True/False')
      else :
        self.hunger = hunger
      if not isinstance(data[3], str) :
        raise Exception ('Boundaries phải là kiểu chuỗi')
      else :
        self.boundaries = boundaries
      if not isinstance(data[4], str) :
        raise Exception ('Location phải là kiểu chuỗi')
      else :
        self.location = location
    except Exception as E :
      print(E)
  @abstractmethod
  def makeNoise() :
    pass
  @abstractmethod
  def eat() :
    pass
  @abstractmethod
  def roam(self):
    pass
  def sleep(self) :
    return random.choice(['Đang ngủ','Đang thức'])
  def __str__(self) :
    return f'Picture: {self.picture}, Food: {self.food}, Hunger: {self.hunger}, Boundaries: {self.boundaries}, Location: {self.location}'
```

```python
class Feline(Animal) :
    def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
        super().__init__(picture, food, hunger, boundaries, location)
    
    def roam(self) :
        return random.choice(['Đang đi dạo','Đang nằm nghỉ'])
```

```python
class Canine(Animal) :
    def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
        super().__init__(picture, food, hunger, boundaries, location)

    def roam(self) :
        return random.choice(['Đang đi dạo','Đang nằm nghỉ'])
```

```python
class Lion(Feline) :
  def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
    super().__init__(picture, food, hunger, boundaries, location)
  def makeNoise(self):
    print("Sư tử đang kêu")
  def eat(self) :
    if self.food == 'Chưa xác định' :
      print('Sư tử đang không ăn gì')
    else :
      print(f'Sư tử đang ăn {self.food}')
  def __str__(self):
    return f'Name: Sư tử, '+ super().__str__()
```

```python
class Cat(Feline) :
  def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
    super().__init__(picture, food, hunger, boundaries, location)
  def makeNoise(self):
    print("Mèo đang kêu")
  def eat(self) :
    if self.food == 'Chưa xác định' :
      print('Mèo đang không ăn gì')
    else :
      print(f'Mèo đang ăn {self.food}')
  def __str__(self):
    return f'Name: Mèo, '+ super().__str__()
```

```python
class Tiger(Feline) :
  def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
    super().__init__(picture, food, hunger, boundaries, location)
  def makeNoise(self):
    print("Hổ đang kêu")
  def eat(self) :
    if self.food == 'Chưa xác định' :
      print('Hổ đang không ăn gì')
    else :
      print(f'Hổ đang ăn {self.food}')
  def __str__(self):
    return f'Name: Hổ, '+ super().__str__()
```

```python
class Hippo(Animal) :
  def __init__(self, picture: None, food: None, hunger: None, boundaries: None, location: None) -> None:
    super().__init__(picture, food, hunger, boundaries, location)
  def makeNoise(self):
    print("Hà mã đang kêu")
  def eat(self) :
    if self.food == 'Chưa xác định' :
      print('Hà mã đang không ăn gì')
    else :
      print(f'Hà mã đang ăn {self.food}')
  def roam(self) :
    return random.choice(['Đang đi dạo','Đang nằm nghỉ'])

  def __str__(self):
    return f'Name: Hà mã, '+ super().__str__()
```

```python
class Person :
    def __init__(self, name : None , birthday : None) -> None:
        try :
            if isinstance(name, str) :
                self.name = name 
            else :
                raise Exception('Name phải là kiểu chuỗi')
            datetime.strptime(birthday, '%d-%m-%Y')
            self.birthday = birthday
        except ValueError :
            print('Ngày không hợp lệ')
        except Exception as E :
            print(E)
    
    def getName(self) :
        return self.name
    
    def __str__(self) -> str:
        return f'Name : {self.name} , Birtday : {self.birthday}'
```

```python
class Employee (Person) :
    def __init__(self, name: None, birthday: None, salary : None) -> None:
        try :
            super().__init__(name, birthday)
            if isinstance(salary,float) or isinstance(salary,int) or salary > 0 :
                self.salary = salary
            else :
                raise Exception('Lương không hợp lệ')
        except Exception as E :
            print(E)

    def getSalary(self) :
        return self.salary

    def __str__(self) -> str:
        return super().__str__() + f', Salary : {self.salary}'
```

```python
class Manager(Employee) :
    def __init__(self, name: None, birthday: None, salary: None, assistant : None) -> None:
        try :
            super().__init__(name, birthday, salary)
            if isinstance(assistant,Employee) :
                self.assistant = assistant
            else :
                raise Exception('Kiểu của assistant không phù hợp')
        except Exception as E:
            print(E)
    
    def setAssistant(self, assistant : None) :
        try :
            if isinstance(assistant,Employee) :
                self.assistant = assistant
            else :
                raise Exception('Kiểu của assistant không phù hợp')
        except Exception as E :
            print(E)
    
    def __str__(self) -> str:
        return super().__str__() + f', Assistant : {self.assistant.name}'
```

```python
# Định nghĩa interface Expression
class Expression:
    def toString(self):
        pass

    def evaluate(self):
        pass

# Định nghĩa interface BinaryExpression kế thừa từ Expression
class BinaryExpression(Expression):
    def left(self):
        pass

    def right(self):
        pass

# Cài đặt lớp Numeral
class Numeral(Expression):
    def __init__(self, value):
        self.value = value

    def toString(self):
        return str(self.value)

    def evaluate(self):
        return self.value

# Cài đặt lớp Square
class Square(Expression):
    def __init__(self, expression):
        self.expression = expression

    def toString(self):
        return f"({self.expression.toString()})^2"

    def evaluate(self):
        return self.expression.evaluate() ** 2

# Cài đặt lớp Addition
class Addition(BinaryExpression):
    def __init__(self, left, right):
        self.left = left
        self.right = right

    def toString(self):
        return f"({self.left.toString()} + {self.right.toString()})"

    def evaluate(self):
        return self.left.evaluate() + self.right.evaluate()

# Ví dụ sử dụng các lớp
num1 = Numeral(3)
num2 = Numeral(4)
sum_expression = Addition(num1, num2)
square_expression = Square(sum_expression)

print(sum_expression.toString())  # Output: (3 + 4)
print(sum_expression.evaluate())  # Output: 7
print(square_expression.toString())  # Output: ((3 + 4))^2
print(square_expression.evaluate())  # Output: 49
```
