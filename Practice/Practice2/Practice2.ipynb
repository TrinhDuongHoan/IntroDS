```python
# Hàm tính giá trị lớn nhất của một cột
def max_value_column(data, column):
    max_value = 0
    for value in data[column] :
        if value > max_value :
            max_value = value
    return max_value
```

```python
# Hàm tính giá trị nhỏ nhất của một cột 
def min_value_column(data, column) :
    min_value = float('inf')
    for value in data[column] :
        if value < min_value :
            min_value = value
    return min_value
```

```python
# Hàm tính giá trị trung bình của một cột
def median_value_column(data, column) :
    sum_value = 0
    for value in data[column] :
        sum_value += value
    return sum_value/len(data[column])
```

Câu 5 : Tính độ lệch chuẩn 

```python
def standard_deviation(data, column):
    # Số lượng phần tử của cột
    n = len(data[column])

    # Tính giá trị trung bình của cột
    median = median_value_column(data, column)

    # Tính tổng bình phương sai số
    sum_squared_diff = sum((x - median) ** 2 for x in data[column])

    standard_deviation = (sum_squared_diff / (n - 1)) ** 0.5
    return standard_deviation
```

```python
def pearson_correlation(data, column1, column2):
    # Kiểm tra điều kiện hai column có cùng kích thước
    if len(data[column1]) != len(data[column2]):
        return None
    
    n = len(data[column1])

    # Tính giá trị trung bình của mỗi column
    median1 = median_value_column(data, column1)
    median2 = median_value_column(data, column2)

    # Tính tử số và mẫu số
    numerator = sum((x - median1) * (y - median2) for x, y in zip(data[column1], data[column2]))
    denominator1 = sum((x - median1) ** 2 for x in data[column1])
    denominator2 = sum((y - median2) ** 2 for y in data[column2])

    # Kiểm tra điều kiện mẫu khác 0
    if denominator1 == 0 or denominator2 == 0:
        return None

    pearson_correlation = numerator / ((denominator1 * denominator2) ** 0.5)
    return pearson_correlation
```

Câu 8

```python
# Mối tương quan của từng cột với cột cuối cùng
correlations = data.iloc[:,:-1].apply(lambda col: pearson_correlation(data, col.name, 'class'))
correlations
```

```python
# Trực quan hóa kết quả bằng biểu đồ
import matplotlib.pyplot as plt 
import seaborn as sns
box = pd.DataFrame(correlations, columns=['class'])
heatmap = sns.heatmap(box,annot=True, cmap = 'coolwarm', fmt = '.3f', linewidths=.5)
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation = 0)
plt.title("Mối tương quan pearson giữa các cột với cột cuối cùng")
plt.show()
```
